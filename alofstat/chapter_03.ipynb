```python
# deps
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Exercise 9:

def aggre_avg(x):
    for i in range(1, x.size):
        x[i] = float(i)/ (i + 1) * x[i - 1] + x[i]
        
    return x;

samples_gauss = np.random.normal(loc=0, scale=1, size=10000)
gauss_avg = aggre_avg(samples_gauss)
h1 = plt.plot(gauss_avg)

samples_cauchy = np.random.standard_cauchy(10000)
cauchy_avg = aggre_avg(samples_cauchy)
h2 = plt.plot(cauchy_avg)

plt.legend(('guassian', 'cauchy'))
plt.show()
```

We can look at the variance of the sum $S_n = n^{-1} \sum X_i$. 

$Var(S_n) = Var(n^{-1} \sum X_i) = n^{-2} \sum Var(X_i)$

For guassian, $Var(S_n) = n^{-2} \rightarrow 0 $ as $n \rightarrow \infty$

For Cauchy, $Var(X) \approx E[X^2] = \int \frac{x^2}{ax^2 + bx + c } dx $, the integrand $\rightarrow 1$ as $x \rightarrow \infty$ hence the integral is divergent. So Cauchy distribution has infinite variance which makes $S_n$ highly unstable.
